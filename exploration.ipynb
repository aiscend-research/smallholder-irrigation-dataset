```python
import pandas as pd

# Load the main labeled data
df = pd.read_csv('data/labels/labeled_surveys/random_sample/latest_irrigation_table.csv')

# See how big it is
print("Shape:", df.shape)

# See the column names
print("\nColumns:", df.columns.tolist())

# See the first few rows
print("\nFirst 5 rows:")
df.head()
```

```python
# How many irrigated vs non-irrigated?
print("=== IRRIGATION COUNTS ===")
print(df['irrigation'].value_counts())

# What years are covered?
print("\n=== YEARS ===")
print(df['year'].value_counts().sort_index())

# Small scale vs industrial
print("\n=== SCALE (small-scale coverage stats) ===")
print(df['percent_coverage_hc_small-scale'].describe())
```

```python
# Check what values irrigation column has
print(df['irrigation'].unique())

# Look at the README to understand the scale
# For now let's see if 1 = no irrigation
print("\n=== Sample of irrigation=1 rows ===")
print(df[df['irrigation']==1][['x','y','year','percent_coverage_hc_small-scale']].head())
```

```python
import matplotlib.pyplot as plt

# Plot irrigation distribution
plt.figure(figsize=(10, 4))

# Chart 1: Irrigation levels
plt.subplot(1, 2, 1)
df['irrigation'].value_counts().sort_index().plot(kind='bar', color='steelblue')
plt.title('Irrigation Level Distribution')
plt.xlabel('Irrigation Level')
plt.ylabel('Count')

# Chart 2: Images per year
plt.subplot(1, 2, 2)
df['year'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title('Images Per Year')
plt.xlabel('Year')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
