```python
import sys, os
from datetime import datetime
import shutil

sys.path.append(os.path.join(os.path.dirname(os.getcwd()), '../../../'))
from src.features.download_sentinel2 import initialize_earthengine, retrieve_time_series_stack, visualize_time_series_stack

initialize_earthengine()
```

```python
# Test output directory
TEST_OUT_DIR = './test_timeseries_output_L1C'
os.makedirs(TEST_OUT_DIR, exist_ok=True)

print("="*70)
print("TEST 1: L1C data")
print("="*70)

file_id="site_123_2023_6_15_L1C"

try:
    retrieve_time_series_stack(
        file_id=file_id,
        lat=37.5,
        lon=-120.5,
        date=datetime(2023, 6, 15),
        collection="L1C",
        out_dir=TEST_OUT_DIR,
        start_month=1,
        num_windows=12,  # Use fewer windows for faster testing
        timestep=10,
        window_buffer=1
    )
    print("\n✓ retrieve_time_series_stack() completed successfully\n")

    visualize_time_series_stack(TEST_OUT_DIR, file_id)

    # delete the time series 
    shutil.rmtree(TEST_OUT_DIR)
            
except Exception as e:
    print(f"\n✗ retrieve_time_series_stack() failed: {e}\n")
    import traceback
    traceback.print_exc()
    sys.exit(1)
```
